```python
def calculate_energy(list_values):
    eng=pow(np.linalg.norm(list_values),2)
    
    return [eng]

def get_eeg_features(dataset, waveletname,level):
    data = dataset.copy()
    eeg_features = []

    for signal_no in range(0, len(dataset)):
            
            for array in range(0,len(dataset)):
                features = []
                signal = dataset[signal_no]
                list_coeff = pywt.wavedec(signal, waveletname, level=level)
                
                for coeff in list_coeff:  
                    features += calculate_energy(coeff)
                    
            eeg_features.append(features)
    X = np.array(eeg_features)
    X = X.flatten()
    X = X.tolist()
    
    return X


def get_labels(number):
    num = ["%02d" % i for i in range(number)]
    num.append("approx")
    num.reverse()
    var = []
    e_num = ["%02d" % i for i in range(1)]
    electrode = 1
    labels=[]
    e_string = "e"
    cols=[]
                

    for e in range(electrode):
        
        for i in num:
            feat = ['energy', ]
            for f in feat:
                var += [ e_string+str(e)+"_"+ f + "_" + str(i) ]   
    var.append('y')
    return var



def remap_y(text):
    value = 0
    if text == "preictal":
        value = 1
    else:
        pass
    return value

def get_corr_labels(electrode): 
    v = []
    labels=[]
    e_string = "e_"
    cols=[]

    for e in range(electrode):
        v += [ e_string+str(e)+"_corr" ]

    return v
        
```

```python
def get_data(file_list):
    start = time.time()
    data_df = pd.DataFrame()
    signal_df = pd.DataFrame()
    dict_eeg_data = defaultdict(list)
    dict_eeg_corr = defaultdict(list)
    original_name = []
    time_index = []
    
    for filename in file_list:
        
        
        
        #get patient name 
        pattern = '^([^_]*_){2}'
        splice = (filename.rsplit('/', 1)[-1])
        patient = (re.match(pattern, splice)).group(0)
        
        if patient[0] == 'P':
            dec_factor = 13
            upper_limit = 11989+11539
        else:
            dec_factor = 1
            upper_limit = 11989+11989

        #get segment type
        pattern = '^([^_]*_){3}'
        splice = (filename.rsplit('/', 1)[-1])
        segment = ((re.match(pattern, splice)).group(1)).replace("_", "")
        y_val = remap_y(segment)
        
        #get file number
        file_number = (filename.split("."))[0][-4:].lstrip("0")
        
        #generate file name
        name = segment + "_segment_" + file_number

        #get recording number 
        recording = (filename.rsplit('/', 1)[-1])[-8:-4]
        
        
        #set number of electrodes to sample from
        electrodes = 1

        #load data using loadmat
        eeg_data = sio.loadmat(filename)

        
        
        
        #create list to hold the data
        eeg_signals = []
       
        #create a range from the number of electrodes to access files
        for i in list(range(electrodes)):
            

            
           
            #access each array by electrode index number
            eeg_signal = signal.resample(((eeg_data[name][0][0][0][i])),153600)
            eeg_signals.append(eeg_signal)
           
            
            
            #decompose signals and get coeff
            eeg_feat = get_eeg_features(eeg_signals, 'db3', 6)
           
            
            
            
            #append segment type to features 
            eeg_feat.append(y_val)
            
              #use file name as recording name 
        file_labels = [str(filename.rsplit('/', 1)[-1])]

        dict_start=time.time()
        
        for label in file_labels:
            #append data from eeg_data to dictionary using file name as label
                dict_eeg_data[label].append(eeg_feat)
                
                
        eeg_signals = []        
        for i in list(range(15)):

            #access each array by electrode index number
            
            eeg_signal = (eeg_data[name][0][0][0][i])
            eeg_signals.append(eeg_signal)
            
            #get correlation coeff
            
            eeg_corr = np.corrcoef(eeg_signals)
           
            
        
        #access each electrode name by electrode index number 
        file_labels = [str(filename.rsplit('/', 1)[-1])]
        #eeg_labels_ = eeg_data[name][0][0][3][0][i]
        for label in file_labels:
            #append data from eeg_data to dictionary using electrode name as label
                dict_eeg_corr[label].append(eeg_corr[0])
                
        #create df from correlations 
        corr_df = pd.DataFrame(dict_eeg_corr).T.reset_index()
        c = pd.concat([corr_df[0].apply(pd.Series)], axis=1)
        corr_cols = get_corr_labels(15)
        c.columns = corr_cols
        c['Recording'] = corr_df['index']
        
                 #reorder columns
        c_cols = list(c.columns)
        c_cols = [c_cols[-1]] + c_cols[:-1]
        c = c[c_cols]
        
        #drop correlation e00 with e00
        c=c.drop('e_0_corr', axis=1)
        
        #create df from signal coeff
        df = pd.DataFrame(dict_eeg_data).T.reset_index()
        x = pd.concat([df[0].apply(pd.Series)], axis=1)
        
        #name signal coeff columns
        column_names=get_labels(6)
        x.columns = column_names
        x['Recording'] = df['index']
        
        
           #reorder columns
        cols = list(x.columns)
        cols = [cols[-1]] + cols[:-1]
        x = x[cols]
        
        #combine df
        results = pd.merge(c,x,on='Recording')
    
    end=time.time()
    print("This took: "+ str(end-start))
    return results


        
```

```python
#create train and val list of files for all subjects
train = []
test = []



#generate list of subjects from dictionary keys 
subjects = sorted(filenames.keys())

for s in subjects:
    tr, ts = train_test_sequence(s,randomseed=33)
    train.extend(tr)
    test.extend(ts)  

    #add directory to each file name 
dir_st = ""
train = [dir_st + s for s in train]
test = [dir_st + s for s in test]


```

```python
#create train df 
train_df = get_data(train)
#save train data 
train_df.to_csv("train.csv", index=False)
```
